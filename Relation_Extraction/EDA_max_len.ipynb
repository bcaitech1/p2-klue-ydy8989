```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_path = '../input/data/train/'
test_path = '../input/data/test/'

train_df = pd.read_csv(train_path+'train.tsv', delimiter='\t', header=None)
test_df = pd.read_csv(test_path+'test.tsv', delimiter='\t', header=None)
```

```python
# train_df[9] 컬럼 초기화
train_df[9] = 0
# 각 문장의 길이를 구해서 넣어주기
for i in range(9000):
    len_df = len(train_df[1][i])
    train_df[9][i] = len_df

# len를 오름차순으로 정렬
len_list = sorted(train_df[9])

# 90%, 95%, 99% 지점의 값 구하기
for i in range(9000):
    if i == 8100: # 90% : 160
        print(len_list[i])
    if i == 8550: # 95% : 190
        print(len_list[i])
    if i == 8910: # 99% : 278
        print(len_list[i])
```

```python
plt.figure(figsize=(30, 10))
ax = sns.countplot(train_df[9])
ax.set_xticklabels([])
ax = plt.axvline(x=100, color='r', linestyle='--', linewidth=3)
ax = plt.axvline(x=160, color='r', linestyle='--', linewidth=3)
ax = plt.axvline(x=190, color='g', linestyle='--', linewidth=3)
ax = plt.axvline(x=278, color='b', linestyle='--', linewidth=3)
plt.show()
print(min(train_df[9]), max(train_df[9])) # 최소 : 15, 최대 : 510
```

```python
# train_df[9] 컬럼 초기화
test_df[9] = 0
# 각 문장의 길이를 구해서 넣어주기
for i in range(1000):
    len_df = len(test_df[1][i])
    test_df[9][i] = len_df

# len를 오름차순으로 정렬
len_list = sorted(test_df[9])

# 90%, 95%, 99% 지점의 값 구하기
for i in range(1000):
    if i == 900: # 90% : 160
        print(len_list[i])
    if i == 950: # 95% : 190
        print(len_list[i])
    if i == 990: # 99% : 278
        print(len_list[i])
```

```python
plt.figure(figsize=(30, 10))
ax = sns.countplot(test_df[9])
ax.set_xticklabels([])
ax = plt.axvline(x=100, color='r', linestyle='--', linewidth=3)
ax = plt.axvline(x=160, color='r', linestyle='--', linewidth=3)
ax = plt.axvline(x=190, color='g', linestyle='--', linewidth=3)
ax = plt.axvline(x=278, color='b', linestyle='--', linewidth=3)
plt.show()
print(min(test_df[9]), max(test_df[9])) # 최소 : 15, 최대 : 510
```
